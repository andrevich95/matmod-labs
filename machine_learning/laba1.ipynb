```python
import numpy as np
import pandas as pd
```

```python
#Start data
data = np.random.uniform(-1,1,90).reshape(30,3)
w = np.random.uniform(-1,1,90).reshape(30,3)
true_val = np.random.randint(2, size=30)
mu = np.array([0.2, 0.4, 0.6])
porog = np.mean(data)
data
```

```python
activation = lambda x, w : w[0]*x[0] or w[1]*x[1] and w[2]*x[2]
```

```python
def binry_perceptron(x, w, lim_val, ty, mu):
    error = np.sum(mu*data)
    while error != 0 :
        y = []
        act = []
        for i in range(len(x)):
            if activation(x[i],w[i]) >= lim_val:
                y.append(1)
                act.append(activation(x[i],w[i]))
            else:
                y.append(0)
                act.append(activation(x[i],w[i]))
            if y[-1] != ty[i]:
                print(w[i])
                choice = np.random.choice(3, 1, replace=False)
                diff = y[-1] - ty[i]
                delta = mu[choice] * diff * x[i]
                w[i] = w[i] - delta 
        error = abs(np.sum(abs(ty - y)))
    return {
        'w1' : w[:,0],
        'w2' : w[:,1],
        'w3' : w[:,2],
        'theta' : lim_val,
        'x1' : x[:,0],
        'x2' : x[:,1],
        'x3' : x[:,2],
        'a' : act,
        'Y' : y,
        'T' : ty,
        'sigma_x1' : x[:,0]*error,
        'sigma_x2' : x[:,0]*error,
        'sigma_x3' : x[:,0]*error,
        'sigma_theta' : lim_val*error
    }
```

```python
pd.DataFrame(binry_perceptron(data,w,porog,true_val,mu))
```
